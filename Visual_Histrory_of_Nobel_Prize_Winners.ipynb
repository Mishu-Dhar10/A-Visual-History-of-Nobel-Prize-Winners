# Add a column 'usa_born_winner' where the value is TRUE if the 'birth_country' is 'united states of america'

```python
nobel_df['usa_born_winner'] = nobel_df['birth_country'] == 'United States of America'
#usa_born['prize'].value_counts()
```

# Adding a decade columns

```python
# Adding a decade columns
print('Adding a decade column')
nobel_df['decade'] = (np.floor(nobel_df['year'] / 10) * 10).astype('int64')
nobel_df.head()
```

# Proportion of usa_born_winners

```python
prop_usa_winners = nobel_df.groupby('decade', as_index = False)['usa_born_winner'].mean()
prop_usa_winners
```

```python
display(prop_usa_winners)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter

sns.set()
plt.rcParams['figure.figsize'] = [11, 7]

# Plotting USA born winners 
ax = sns.lineplot(x='decade', y='usa_born_winner', data=prop_usa_winners)

# Adding %-formatting to the y-axis
ax.yaxis.set_major_formatter(PercentFormatter(1.0))
```

```python
fig = px.line(data_frame = prop_usa_winners, x = 'decade', y = 'usa_born_winner')
fig.update_layout(yaxis=dict(tickformat='.0%'))
fig.show()
```

# Converting 'birth_date' into dateTime format ( TimeStamp)

```python
nobel_df['birth_date'] = pd.to_datetime(nobel_df['birth_date'])
nobel_df['birth_date'].min()
```

# Calculating the age of the winner
to extract yeaf from a yy-mm-dd column data_fame['column'].dt.year

```python
nobel_df['age'] = nobel_df['year'] - nobel_df['birth_date'].dt.year
nobel_df.head()
```

```python
sns.lmplot(data = nobel_df, x = 'year', y = 'age', lowess = True, line_kws = {'color':'black'})
plt.show()
```

The plot tells a lot. The average age used to be around 55 when they received nobel prize but nowaday it is around 65. But, there is a large spreed in Laureates' ages.We also see that the density is much higher than early 1960s, that means many people share the proce rather the individual winners. There is a disruption between 1939 to 1945.

```python
sns.jointplot(data = nobel_df, x = 'year', y = 'age',
             kind = 'hex')
plt.show()
```

```python
sns.set_style('whitegrid')
# Build a JointGrid comparing humidity and total_rentals
g = sns.JointGrid( x = 'age',
                 y = 'year',
                 data = nobel_df,
                 xlim =(0, 100))
g.plot(sns.regplot, sns.histplot)
plt.show()
```

```python
g = sns.JointGrid(data = nobel_df, x = 'year', y = 'age')
g = g.plot_joint(sns.kdeplot)
g = g.plot_marginals(sns.kdeplot, shade = True)
```

```python
g = (sns.jointplot(x = 'age',
                  y = 'year',
                  kind = 'scatter',
                  xlim = (0, 25000),
                  data = nobel_df).plot_joint(sns.kdeplot)
```

```python
g = (sns.jointplot(x = 'age', y = 'year',
                  data = nobel_df, kind = 'scatter',
                  xlim = (0,100))).plot_joint(sns.kdeplot)
```

# By adding aspect = 2 is its possible to make the plot size bigger.

```python
sns.lmplot(data = nobel_df, x = 'year', y = 'age', lowess = True,
          line_kws = {'color':'black'},
          row = 'category', aspect = 2)
plt.show()
```

We can see the the trend of the plot avobe. In Physics and chemistry the average age is increasing. For the Physics the trend is more clearer. The average age was below 50 but nowadyas it is nearly 70. In Economics, it is more stable. But, peace showing a opposite trend. The avg age of nobel peace proze winners are getting younger.
